```python
import os
from dotenv import load_dotenv
from web3 import Web3

# Load environment variables from .env file
load_dotenv()

# Get private key from environment variables
nft_private_key = os.getenv("NFT_WALLET_PRIVATE_KEY")
test_private_key = os.getenv("TEST_WALLET_PRIVATE_KEY")

alchemy_api_key = os.getenv("ALCHEMY_API_KEY")
if not nft_private_key:
    raise ValueError("Private key not found in .env file. Please add NFT_WALLET_PRIVATE_KEY=your_private_key")
if not test_private_key:
    raise ValueError("Private key not found in .env file. Please add TEST_WALLET_PRIVATE_KEY=your_private_key")
# Connect to blockchain network (using Ethereum as example)
# You can change the provider URL based on which network you want to connect to
provider_url = os.getenv("PROVIDER_URL", f"https://eth-sepolia.g.alchemy.com/v2/{alchemy_api_key}")
w3 = Web3(Web3.HTTPProvider(provider_url))

# Check connection
if w3.is_connected():
    print("Connected to blockchain network")
else:
    print("Failed to connect to blockchain network")

# Create account from private key
nft_account = w3.eth.account.from_key(nft_private_key)
nft_wallet_address = nft_account.address

test_account = w3.eth.account.from_key(test_private_key)
test_wallet_address = test_account.address


print(f"Connected nft wallet address: {nft_wallet_address}")
print(f"Connected test wallet address: {test_wallet_address}")
```

Connect to the NFT contract

```python
# Contract address on Sepolia testnet
contract_address = '0xf18E3901D91D8a08380E37A466E6F7f6AA4BD4a6'


# Create contract instance
nft_contract = w3.eth.contract(address=contract_address, abi=abi)
```

what is the mint price ?

```python
# Mintpreis in Wei abrufen
mint_price_wei = nft_contract.functions.mintPrice().call()

# Konvertiere von Wei zu ETH fr bessere Lesbarkeit
mint_price_eth = w3.from_wei(mint_price_wei, 'ether')

print(f"Mintpreis: {mint_price_wei} Wei ({mint_price_eth} ETH)")
```

```python

# Definiere Token URI - verwende eine echte URL statt leerem String
token_uri = "https://example.com/metadata/test-nft.json"

# Definiere mint_price (0.01 ETH in Wei)
mint_price = w3.to_wei(mint_price_eth, 'ether')

# Vereinfachte Transaktion
try:
    
    tx_hash = nft_contract.functions.safeMint(token_uri).transact({
        'from': test_wallet_address,
        'value': mint_price_wei
    })
    
    # Warte auf Besttigung
    tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
    print(f"NFT erfolgreich geminted! Tx: {tx_hash.hex()}")
    
except Exception as e:
    print(f"Fehler beim Minten: {e}")
```
