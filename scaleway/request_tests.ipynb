```python
import requests
```

```python
response = requests.get("http://localhost:8080", params={"prompt": "The picture of Frankfurt Bahnhofsviertel at night. It should be done in comic style."})
print(f"The response: {response.json()}")
```

```python
from IPython import display
from base64 import b64decode

display.Image(b64decode(response.json()['data'][0]['b64_json']))
```
